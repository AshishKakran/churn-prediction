## Data preparation
Following data preparation steps are required
* Filling NaN values across numerical columns
* Encoding categorical columns to One-hot type value
* Filling Nan values across categorical columns
* Scale features

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import  OneHotEncoder,StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
def label_splitter(frame):
    return frame['churn'], frame.drop('churn',axis=1)
```

```python
d_label,d_features=label_splitter(d)
```

```python
d_features.shape
```

```python
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")), #fill NaN values in numerical columns by median of column
    ('std_scaler', StandardScaler()),
])

cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="most_frequent")), # fill NaN values in categorical columns by mode of column
    ('encoder', OneHotEncoder(sparse=False)) #sparse=False
])

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_features),
    ("cat", cat_pipeline, cat_features)
])
```

```python
d_prep=full_pipeline.fit_transform(d_features)
```

```python
d_prep.shape
```

```python
train_X=d_prep[:80000,]; test_X = d_prep[80000:,]; train_y = d_label[:80000].values;test_y=d_label[80000:].values #a bad way to split dataset
```

## Modelling

 In telecom context we can have following interpretation of metrics:
 * **High precision**: It means model is able to classify people likely to churn correctly with high accuracy
 * **High recall**: It means model is able to positively classify most of the people likely to churn
 * **Low precision**: It means model is overkill and positively classify some people who are not likely to churn
 * **Low recall** : It means model is underperforming and doesn't classify some people who are likely to churn
 
  **Thus we prefer a model high precision and high recall.**

Following models are trained and compared
 * Logistic Regressor
 * Random forest Classifier
 * SVM regressor

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.metrics import roc_curve
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
metrics = {}
```

```python
def model_metrics(modelName, y_label,y_pred):
    print(modelName)
    precision = precision_score(y_label,y_pred)
    print("Precision: ", precision)
    recall = recall_score(y_label,y_pred)
    print("Recall: ", recall)
    f1 = f1_score(y_label,y_pred)
    print("F1 score: ", f1)
    
    metrics[modelName] = {'Precision': precision, 'Recall': recall, 'F1_score': f1}
    
    cm = confusion_matrix(y_label, y_pred)
    cmd_obj = ConfusionMatrixDisplay(cm, display_labels=['churn', 'notChurn'])
    cmd_obj.plot()
    cmd_obj.ax_.set(
                    title='Sklearn Confusion Matrix with labels!!', 
                    xlabel='Predicted Churn', 
                    ylabel='Actual Churn')
    save_fig(modelName + "CM")
    plt.show()
    
    
```

```python
def plot_roc_curve(fpr,tpr, label=None):
    plt.plot(fpr,tpr,linewidth=2,label=label)
    plt.plot([0,1],[0,1], 'k--') 
    plt.axis([0,1,0,1])
    plt.xlabel('False Positive Rate(precision)', fontsize=16)
    plt.ylabel('True Positive Rate(recall)', fontsize=16)
    plt.grid(True)
```

### Model 1:  Lasso Logistic Regressor

* Default version of LogisticRegression is Lasso Regression in which L2 norm is penalised

```python
from sklearn.linear_model import LogisticRegression
Lasso_clf = LogisticRegression(random_state=42, tol=0.01,solver='newton-cg', max_iter=200) # by default penalty is l2
```

```python
cross_val_score(Lasso_clf, train_X, train_y, cv=3, scoring="accuracy")
```

```python
y_pred = cross_val_predict(Lasso_clf, train_X, train_y, cv=3)
model_metrics("Lasso_Logistic_Regressor", train_y,y_pred)
```

```python
y_scores = cross_val_predict(Lasso_clf,train_X, train_y, cv=3, method="decision_function")
fpr1,tpr1,thresholds = roc_curve(train_y,y_scores) # will be used later to plot ROC curve
```
